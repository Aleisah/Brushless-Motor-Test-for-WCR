```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Motor_test.csv')
df = pd.DataFrame(data)
df
```

```python
df.shape
```

```python
# Visualization

fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Throttle vs Thrust
axs[0].plot(df["Throttle (0-1)"], df["Thrust (g)"], marker='o', label="Throttle vs Thrust")
axs[0].set_xlabel("Throttle (0-1)")
axs[0].set_ylabel("Thrust (g)")
axs[0].set_title("Throttle vs Thrust")
axs[0].grid()
axs[0].legend()

# Throttle vs Speed
axs[1].plot(df["Throttle (0-1)"], df["Speed (RPM)"], marker='o', label="Throttle vs Speed", color="orange")
axs[1].set_xlabel("Throttle (0-1)")
axs[1].set_ylabel("Speed (RPM)")
axs[1].set_title("Throttle vs Speed")
axs[1].grid()
axs[1].legend()

# Throttle vs Temperature
axs[2].plot(df["Throttle (0-1)"], df["Temp (C)"], marker='o', label="Throttle vs Temp", color="green")
axs[2].set_xlabel("Throttle (0-1)")
axs[2].set_ylabel("Temp (C)")
axs[2].set_title("Throttle vs Temp")
axs[2].grid()
axs[2].legend()

plt.tight_layout()
plt.show()
```

```python
# Best-Fit Line (Linear Regression) for Throttle vs Thrust

X_throttle = df["Throttle (0-1)"].values.reshape(-1, 1)
y_thrust = df["Thrust (g)"].values

model_thrust = LinearRegression()
model_thrust.fit(X_throttle, y_thrust)

# Equation for Throttle vs Thrust
thrust_intercept = model_thrust.intercept_
thrust_slope = model_thrust.coef_[0]
print(f"Throttle vs Thrust Equation: y = {thrust_slope:.2f}x + {thrust_intercept:.2f}")

# Best-Fit Line (Linear Regression) for Throttle vs Speed
y_speed = df["Speed (RPM)"].values

model_speed = LinearRegression()
model_speed.fit(X_throttle, y_speed)

# Equation for Throttle vs Speed
speed_intercept = model_speed.intercept_
speed_slope = model_speed.coef_[0]
print(f"Throttle vs Speed Equation: y = {speed_slope:.2f}x + {speed_intercept:.2f}")

# Plot the Best-Fit Lines
plt.figure(figsize=(12, 6))

# Throttle vs Thrust with Best-Fit Line
plt.subplot(1, 2, 1)
plt.scatter(df["Throttle (0-1)"], df["Thrust (g)"], color="blue", label="Data")
plt.plot(df["Throttle (0-1)"], model_thrust.predict(X_throttle), color="red", label="Best-Fit Line")
plt.xlabel("Throttle (0-1)")
plt.ylabel("Thrust (g)")
plt.title("Throttle vs Thrust with Best-Fit Line")
plt.legend()
plt.grid()

# Throttle vs Speed with Best-Fit Line
plt.subplot(1, 2, 2)
plt.scatter(df["Throttle (0-1)"], df["Speed (RPM)"], color="orange", label="Data")
plt.plot(df["Throttle (0-1)"], model_speed.predict(X_throttle), color="red", label="Best-Fit Line")
plt.xlabel("Throttle (0-1)")
plt.ylabel("Speed (RPM)")
plt.title("Throttle vs Speed with Best-Fit Line")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
